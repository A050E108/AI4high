# Matplotlib範例學習快速入門

### 範例一:簡單圖形
```
步驟一:先產生x軸的資料===使用陣列:0到2π之間，以0.01為step
t = np.arange(0.0, 2.0*np.pi, 0.01)	

步驟二:針對每一個x產生 y (y = sin(x))==== y 也是一個陣列

s = np.sin(t)			
```

````
import numpy as np
import pylab as pl


# 產生資料
t = np.arange(0.0, 2.0*np.pi, 0.01)	
s = np.sin(t)			

#畫圖

pl.plot(t,s)		
pl.xlabel('x')			
pl.ylabel('y')
pl.title('sin')		
pl.show()				
````

使用plot()函數畫圖

```
pl.plot(t,s)			#畫圖，以t為橫坐標，s為縱坐標
pl.xlabel('x')			#設定坐標軸標籤
pl.ylabel('y')
pl.title('sin')		#設定圖形標題
pl.show()				#顯示圖形
```

### 範例二:使用scatter()圖來顯示資料的關係

````
import numpy as np
import pylab as pl

# 產生資料
x = np.arange(0, 2.0*np.pi, 0.1)
y = np.cos(x)

#畫圖
pl.scatter(x,y)			
pl.show()
````

### 範例三:使用scatter()圖::你看得出資料有何關係??

s指大小，c指顏色，marker指符號形狀

````
import matplotlib.pylab as pl
import numpy as np

# 產生資料
x = np.random.random(100)
y = np.random.random(100)

#畫圖
pl.scatter(x,y,s=x*500,c=u'r',marker=u'*')	
pl.show()
````

### 範例四:

````
import numpy as np
import pylab as pl
import matplotlib.font_manager as fm

# 產生資料
x = np.arange(0.0, 2.0*np.pi, 0.01)
y = np.sin(x)						
z = np.cos(x)						

#畫圖
pl.plot(x, y, label='sin()')
pl.plot(x, z, label='cos()')
pl.xlabel('x')		
pl.ylabel('y')
pl.title('sin-cos') 
pl.legend()				
pl.show()
````

### 範例五:  帶有數學公式的圖形

````
import numpy as np
import matplotlib.pyplot as plt


# 產生資料
x = np.linspace(0, 2*np.pi, 500)
y = np.sin(x)
z = np.cos(x*x)


#畫圖
plt.figure(figsize=(8,5))
#標籤前後加上$，代表以內嵌的LaTex引擎顯示為公式
plt.plot(x,y,label='$sin(x)$',color='red',linewidth=2)	#紅色，2個像素寬
plt.plot(x,z,'b--',label='$cos(x^2)$')		#b::藍色，--::虛線
plt.xlabel('Time(s)')
plt.ylabel('Volt')
plt.title('Sin and Cos figure using pyplot')
plt.ylim(-1.2,1.2)

plt.legend()							
plt.show()
````

### 範例六:多圖並呈===使用subplot(函數)

````
import numpy as np
import matplotlib.pyplot as plt

# 產生資料
x= np.linspace(0, 2*np.pi, 500)	
y1 = np.sin(x)					
y2 = np.cos(x)
y3 = np.sin(x*x)

#畫圖
plt.figure(1)					#建立圖形
#create three axes
ax1 = plt.subplot(2,2,1)			#第一列第一行圖形
ax2 = plt.subplot(2,2,2)			#第一列第二行圖形
ax3 = plt.subplot(2,1,2)			#第二列

##設定第一張圖
plt.sca(ax1)					#選擇ax1
plt.plot(x,y1,color='red')		#繪製紅色曲線
plt.ylim(-1.2,1.2)				#限制y坐標軸範圍

##設定第二張圖
plt.sca(ax2)					#選擇ax2
plt.plot(x,y2,'b--')			#繪製藍色曲線
plt.ylim(-1.2,1.2)

##設定第三張圖
plt.sca(ax3)					#選擇ax3
plt.plot(x,y3,'g--')
plt.ylim(-1.2,1.2)

##多圖並呈
plt.show()
````

### 範例七:

````
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d


# 產生資料
x,y = np.mgrid[-2:2:20j, -2:2:20j]
z = 50 * np.sin(x+y)		#測試資料

#畫圖
ax = plt.subplot(111, projection='3d')	#三維圖形
ax.plot_surface(x,y,z,rstride=2, cstride=1, cmap=plt.cm.Blues_r)
ax.set_xlabel('X')			#設定坐標軸標籤
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plt.show()
````

### 範例八:

````
import pylab as pl
import numpy as np
import mpl_toolkits.mplot3d

rho, theta = np.mgrid[0:1:40j, 0:2*np.pi:40j]
z = rho**2
x = rho*np.cos(theta)
y = rho*np.sin(theta)
ax = pl.subplot(111, projection='3d')
ax.plot_surface(x,y,z)
pl.show()
````
